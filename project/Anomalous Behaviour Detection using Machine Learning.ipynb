<span style="color:#000; font-family: 'Bebas Neue'; font-size: 1.3em;"> 1. Data sets with anomalous and normal behaviour 
2. Develop a deep learning model to make decision </span>


<span style="color:#000; font-family: 'Bebas Neue'; font-size: 1.3em;"> We have gathered data sets which includes data packets with anomalous as well as normal behavior ,53.5 % data is abnormal and 46.5 % data is normal data. We have studied the parameters which makes drastic effect in decision making process . As per our datasets we consider as major parameters the ones listed below. The percentage listed next to each one of these is a percentage of effectiveness provided by the previous study performed at Aberystwyth in the UK. 

    Parameters                     Weight ( Effect in % ) 
    i) Index                       29 %
